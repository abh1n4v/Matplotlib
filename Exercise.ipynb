```python
#importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('company_sales_data.csv')
df.head()
```

```python
#1
plt.plot(df['month_number'], df['total_profit'])
plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company profit per month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()

```

```python
#2

plt.plot(df['month_number'], df['total_profit'], 'ro--', linewidth=3, markerfacecolor= 'black', label= 'profit data last year' )
plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company profit per month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc=4)
plt.show()

```

```python

#3

for sales_data in df.columns[1:7]:
    plt.title("Sales Data")
    plt.xlabel('Month Number')
    plt.ylabel('Sales units in number')
    plt.xticks(df['month_number'])
    plt.plot(df['month_number'],df[sales_data], label=sales_data, marker= 'o')
    plt.legend(loc="upper left")

plt.plot(df['month_number'],df[sales_data], label=sales_data, marker= 'o')

```

```python
#4
plt.figure(figsize=(8,5))
plt.scatter(df['month_number'],df['toothpaste'])
plt.title('Toothpaste Sales Data')
plt.ylabel('No. of units sold')
plt.xlabel('Month number')
plt.grid(True,linewidth= 2, linestyle='--')
plt.show()
```

```python
#5
df.plot(x='month_number', y=['facecream','facewash'],label=['Face Cream', 'Face Wash'], kind='bar')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.grid(True,linestyle='--')

#6 save part
plt.savefig('Question 6.png',dpi=300)
```
