```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))

plt.plot(gas['Year'],gas['USA'],'b.-',label='USA')
plt.plot(gas.Year,gas.Canada,'g.-',label='Canada')
plt.plot(gas.Year,gas.Australia, 'r.-', label='Aus')

#for country in gas:
#    if country !=  'Year': #to skip Year column
#        plt.plot(gas['Year'],gas[country],marker='.')
plt.title('Gas Prices over Time')
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.xticks(gas['Year'][::3]) #skips every 3 years
plt.legend()
plt.show()
```

Weights of Fifa Players

```python
fifa['Weight'] = [int(x.strip('lbs')) if type(x)==str else x for x in fifa['Weight']]
fifa['WeightInKg'] = (fifa['Weight']/2.205).round(2)

plt.style.use('ggplot')
light = fifa.loc[(fifa['WeightInKg'] <70)].count()[0]
medium = fifa.loc[(fifa['WeightInKg'] >=70) & (fifa['WeightInKg'] <80)].count()[0]
heavy = fifa.loc[(fifa['WeightInKg'] >=80)].count()[0]

weights = [light, medium, heavy]
labels = ['<70' , '70-80', '>80']
plt.title("Weight in KG")
plt.pie(weights, labels= labels, autopct= '%.2f%%', pctdistance=0.8, explode=[0,0,0.1])
plt.show()
```
